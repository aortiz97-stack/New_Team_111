```python
import pandas as pd
```

```python
def create_df(file_name, skip_rows = [0]):
    """
    This file creates a data frame from reading a provided `file_name`.
    
    Arguments:
    `file_name`: name of the file, string
    `skip_rows`: array, tells which rows to skip when reading a csv file
    
    Output:
    `df`: a dataframe containing `file_name` contents
    """
    path = 'Data/' + file_name
    df = pd.read_csv(path, skiprows = skip_rows)
    return df
```

```python
def rename_df(df):
    """
    This function renames the columns in a dataframe by converting original column names to lowercase,
    removing special characters, and adding a descriptor to the original column name to identify the dataset it came from.
    
    Arguments:
    
    Output:
    `df`: a dataframe with the new column names
    """
    
    column_names = {}
    for name in df.columns:
        column_names[name] = name.replace(",","").replace(" - ","_").replace(" ","_").replace("%","").lower()
    df = df.rename(columns=column_names)
    return df
```

```python
def insert_nan(descriptor):
    if descriptor =='See Plan Brochure':
        return 'NaN'
    return descriptor
```

```python
def string_to_float(descriptor):
    descriptor = (str(descriptor).replace("$","").replace(",","").replace("%",""))
    if descriptor != 'See Plan Brochure':
        descriptor = float(descriptor)
    return descriptor
```
