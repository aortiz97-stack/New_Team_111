# The first step to take in our EDA was to determine what outliers exist, if any. The code below filters out any outliers within each state's data. No outliers were detected in this scenario, nationwide.

# Source: https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba

```python
national_df = df.copy()
indep_features_list = df.drop(columns=['uninsured', 'uninsured_adults', 'uninsured_children']).columns.to_list()
print(national_df.shape)


Q1 = national_df[indep_features_list].quantile(0.25)
Q3 = national_df[indep_features_list].quantile(0.75)
IQR = Q3 - Q1
    
national_df[indep_features_list] = national_df[indep_features_list]\
[~((national_df[indep_features_list]< (Q1 - 1.5 * IQR)) |\
(national_df[indep_features_list] > (Q3 + 1.5 * IQR))).any(axis=1)]

national_df.shape


```

The next step to take in our EDA was to determine what outliers exist, if any. The code below filters out any outliers within each state's data. The sample data was at the state level, not nationwide. No outliers were detected in this scenario.

 Source: https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba

```python
# Looking to see if there are any outliers, looking at each state
print(df.shape)
indep_features_list = df.drop(columns=['uninsured', 'uninsured_adults', 'uninsured_children']).columns.to_list()

for state in df.state_abbreviation.unique():
    Q1 = df[df.state_abbreviation==state][indep_features_list].quantile(0.25)
    Q3 = df[df.state_abbreviation==state][indep_features_list].quantile(0.75)
    IQR = Q3 - Q1
    
    df[df.state_abbreviation==state][indep_features_list] = df[df.state_abbreviation==state][indep_features_list]\
    [~((df[df.state_abbreviation==state][indep_features_list]< (Q1 - 1.5 * IQR)) |\
       (df[df.state_abbreviation==state][indep_features_list] > (Q3 + 1.5 * IQR))).any(axis=1)]


print(df.shape)


```

# There are three independent variables of interest when it comes to looking at uninsurance rates across the nation: teen_births, childhood_immunizations, and percent_non_hispanic_white. Further exploration by clustering counties across the nation according to a low, medium, or high presence of each variable may reveal interesting insights. 

# Before doing that, however, we must see whether the national df is normally distributed, and if the samples' variances are approximately equal to each other. This is in order to see if it is possible to perform an ANOVA test and compare significant differences between low, medium, and high presences for the three variables: teen_births, childhood_immunizations, and percent_non_hispanic_white.

# Because the data is not normally distributed, it is best to use the Kruskal Wallis test instead.

Source: https://www.statology.org/kruskal-wallis-test-python/

```python
# Checking to see how skewed the df is
def create_histo(df, column_name):
    plt.hist(df[column_name])
    
    plt.ylabel("Frequency")
    plt.xlabel(column_name)
    plt.title(column_name.capitalize() + " distribution")
    plt.show()
    
    plt.savefig(('EDA_Visualizations/Histograms/'+ column_name + '_histogram.jpg'))
    plt.close()
```

```python
create_histo(df, 'teen_births')
create_histo(df, 'childhood_immunizations')
create_histo(df, '%_non_hispanic_white')
```

# Starting off with teen_births variable, we can see that there is a significant difference between low, medium, and high teen births' median values for the uninsurance rate. The p-value is about 1.018 e-115. 

```python
df.teen_births.describe()
```

```python
low_threshold = (df.teen_births.max() - 0)/3
high_threshold = (df.teen_births.max() - 0) * 2/3

teen_birth_low_df = df[df.teen_births < low_threshold]
teen_birth_med_df = df[(df.teen_births >= low_threshold) & (df.teen_births <= high_threshold)]
teen_birth_high_df = df[df.teen_births > high_threshold]

# Perform the Kruskal-Wallis test for uninsured columns. 
scipy.stats.kruskal(teen_birth_low_df.uninsured, teen_birth_med_df.uninsured, teen_birth_high_df.uninsured)
```

# There is also statistical difference between low and high rates of childhood immunization rates, when it comes to uninsurance rates.

```python
# Doing the same for childhood_immunizations
df.childhood_immunizations.describe()
```

```python
# Because there are no values less than 0.33, the data will be split only into two groups instead of three. 
child_immun_low_df = df[df.childhood_immunizations < 0.5] # Because this df is empty, will not include

child_immun_high_df = df[df.childhood_immunizations >=0.5]

# Perform the Kruskal-Wallis test for uninsured columns. 
scipy.stats.kruskal(child_immun_low_df.uninsured, child_immun_high_df.uninsured)
```

# The results for different percentage of white population's uninsurance rates is inconclusive looking at low, medium, and high. If we only look at medium and high results, however, we do get significant differences for the uninsurance rate. 

```python
# Doing the same for %_non_hispanic_white
df['%_non_hispanic_white'].describe()
```

```python
low_threshold = 0.330
high_threshold = 0.660

percent_white_low_df = df[df['%_non_hispanic_white'] < low_threshold]
percent_white_med_df = df[(df['%_non_hispanic_white'] >= low_threshold) & (df['%_non_hispanic_white'] <= high_threshold)]
percent_white_high_df = df[df['%_non_hispanic_white'] > high_threshold]

# Perform the Kruskal-Wallis test for uninsured columns. 
print(scipy.stats.kruskal(percent_white_low_df.uninsured, percent_white_med_df.uninsured, percent_white_high_df.uninsured))

# Because the Kruskal-Wallis test above returned an nan p-value for these three categories, we perform the same analysis,
# except we now exclude the low category

scipy.stats.kruskal(percent_white_med_df.uninsured, percent_white_high_df.uninsured)
```

# There is significant difference for low, medium, and high groups' uninsurance rates for the variables of teen births, childhood immunizations, and percentage of nonhispanic whites. The bar charts display these differences for those groups 

```python
def create_bar(df_list, df_list_names):
    """
    Creates bar graph from a list of data frames' values. Df_list and df_list_names must have their respective 
    elements in the same order as each other.
    
    df_list: a list of df
    df_list_names: a list of strings containing the df's names
    """
    x = []
    y = []
    for i in range(len(df_list)):
        x.append(df_list_names[i].split('_df')[0])
        y.append(df_list[i].uninsured.median() * 100.0)
    
    plt.bar(x, y)
    plt.xlabel("Category")
    plt.ylabel("Median Uninsurance Percentage")
    plt.title("Uninsurance Rate Differences Amongst Counties Grouped By Category")
    plt.show()
    plt.savefig("EDA_Visualizations/Bar_Charts/{}_categories".format(df_list_names[0].split("_")[:2]))
    plt.close()

teen_birth_dfs = [teen_birth_low_df, teen_birth_med_df, teen_birth_high_df]
teen_birth_names = ['teen_birth_low_df', 'teen_birth_med_df', 'teen_birth_high_df']
create_bar(teen_birth_dfs, teen_birth_names)

child_immun_dfs = [child_immun_low_df, child_immun_high_df]
child_immun_names = ['child_immun_low_df', 'child_immun_high_df']
create_bar(child_immun_dfs, child_immun_names)

percent_white_dfs = [percent_white_med_df, percent_white_high_df]
percent_white_names = ['percent_white_med_df', 'percent_white_high_df']
create_bar(percent_white_dfs, percent_white_names)
```

# Now, we are going to repeat the same process, except looking at each state instead of the entire US. For this, we will create a dictionary that has the state abbreviation as key, and another dictionary containing the independent variable and respective r-value as the value.



```python
state_corr_dict = defaultdict(dict)
for state in df.state_abbreviation.unique():
    state_corr = df[df.state_abbreviation==state].corr()
    print("State of " + state)
    for index in state_corr.index:
        fltr = (state_corr.loc[index]["uninsured"] >= 0.5) | (state_corr.loc[index]["uninsured"] <= -0.5)
        if fltr and index not in ['uninsured', 'uninsured_adults', 'uninsured_children']:
            print(index)
            print(state_corr.loc[index]["uninsured"][fltr])
            
            state_corr_dict[state][index] = state_corr.loc[index]["uninsured"][fltr]
    print("\n")
```

```python
for state in df.state_abbreviation.unique():
    if (state != 'US') and (state != 'DC'):
        print(state)
        strong_state_corr = df[state_corr_dict[state].keys()].corr()
        plt.title("State of " + state)
        
        print(sns.heatmap(strong_state_corr, cmap='vlag'))
        plt.savefig('EDA_Visualizations/Heatmaps/{}_indep_features_heatmap'.format(state), transparent=True,
                    bbox_inches='tight')
        plt.close()
```

```python

def create_msno(df, start_index, final_index):
    plt.close()
    for state in df.state_abbreviation.unique()[start_index:final_index]:   
        corr = df[df.state_abbreviation==state].drop(columns=['state_fips_code', 'county_fips_code', '5_digit_fips_code', 'uninsured', 'uninsured_adults',
                                'uninsured_children']).corr()
        corr = corr[(corr > -0.7) & (corr < 0.7)]
        (msn.matrix(corr))

        plt.title("State of " + state)


create_msno(df, 0, 20)
create_msno(df, 20, 40)
create_msno(df, 40, 52)

            
    
    
    
```

```python
# Computing Variance Inflation Factors
sub_df = df[df.state_abbreviation=='TX'].drop(columns=['state_fips_code', 'county_fips_code', '5_digit_fips_code',
                                                       'state_abbreviation', 'name','uninsured', 'uninsured_children',
                                                       'uninsured_adults'])
vif_data_copy = sub_df.copy()


while not vif_data_copy.empty:
    interesting_vars = sub_df.columns.to_list()
    X = sub_df[interesting_vars]
    print(X)

    # VIF dataframe
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns

    # calculating VIF for each feature
    
    vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                              for i in range(len(X.columns))]

    print(vif_data)
    break

    if float('inf') in vif_data["VIF"].unique():
        feature_list = vif_data[vif_data['VIF']==float("inf")].feature.to_list()
        sub_df = sub_df.drop(columns=feature_list[0])
        vif_data_copy = vif_data[vif_data.feature!= feature_list[0]]
    elif ~ vif_data[vif_data['VIF'] >= 5].empty:
        feature_list = vif_data[vif_data['VIF']>= 5].feature.to_list()
        sub_df = sub_df.drop(columns=feature_list[0])
        vif_data_copy = vif_data[vif_data.feature!= feature_list[0]]
    else:
        vif_data_copy = pd.DataFrame()

sub_df
```
