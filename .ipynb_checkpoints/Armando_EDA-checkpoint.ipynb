```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
df = pd.read_csv('Data/county_data_clean.csv').drop(columns='Unnamed: 0')
```

```python
df
```

The first step to take in our EDA was to determine what outliers exist, if any. The first round of outlier detection is looking at univariate outliers for each of the features using boxplots (so, not taking into account uninsured, the dependent variable).

```python
indep_features_list = df.drop(columns=['uninsured']).columns.to_list()

for state in df.state_abbreviation.unique():
    #temp_df = df[df.state_abbreviation==state][indep_features_list]
    Q1 = df[df.state_abbreviation==state][indep_features_list].quantile(0.25)
    Q3 = df[df.state_abbreviation==state][indep_features_list].quantile(0.75)
    IQR = Q3 - Q1
    
    df[df.state_abbreviation==state][indep_features_list] = df[df.state_abbreviation==state][indep_features_list]\
    [~((df[df.state_abbreviation==state][indep_features_list]< (Q1 - 1.5 * IQR)) |\
       (df[df.state_abbreviation==state][indep_features_list] > (Q3 + 1.5 * IQR))).any(axis=1)]
    print("state of " + state, df[df.state_abbreviation==state][indep_features_list])



```

```python
# Find and filter out univariate outliers
z_scores = scipy.stats.zscore(df)
abs_z = numpy.abs(z_scores)
fltr = (abs_z < 3).all(axis=1)

df[fltr]
```

```python
# Find univariate outliers for features using boxplots 
indep_features_df = df.drop(columns=['state_fips_code',
 'county_fips_code',
 '5_digit_fips_code',
 'name', 'uninsured'])

def create_boxplot(state, column, df):
    sns.boxplot(x=df[df.state_abbreviation==state][column], orient="h")
    y = np.arange(1)
    
    sns.despine(left=True)
    plt.yticks(y, " ")
    plt.xlabel(column)
    
    plt.title(("State of {0}: {1} Boxplot").format(state, column))
    
    
    plt.savefig(('EDA_Visualizations/Boxplots/{0}_{1}_boxplot.jpg').format(state, column))
    plt.close()
    
for state in indep_features_df.state_abbreviation.unique():
    if state != 'US':
        for column in indep_features_df.columns:
            if column != 'state_abbreviation':
                create_boxplot(state, column, indep_features_df)



```

```python
# Find Outliers 
for state in df.state_abbreviation.unique().tolist():
    temp_df = df[df.state_abbreviation==state]['uninsured']
    Q1 = temp_df.quantile(0.25)
    Q3 = temp_df.quantile(0.75)
    IQR = Q3 - Q1
    
    temp_df_out = temp_df[~((temp_df < (Q1 - 1.5 * IQR)) |(temp_df > (Q3 + 1.5 * IQR))).any(axis=1)]
    df[df.state_abbreviation==state] = temp_df_out

#df = df.dropna(subset='state_abbreviation')
    
df
    
```
