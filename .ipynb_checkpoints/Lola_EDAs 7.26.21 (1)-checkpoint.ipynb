```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("county_data_clean.csv")
data.head(3)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df=data.iloc[:,5:]
df.groupby('name').mean()
df=df[df.name!='United States']   ###United States is not a couty.
#df=df.fillna(df.mean(numeric_only=True))  ### when you fill nan values with column mean, then the correlation values will change. Same if you normalize the dataframe.
df.head(3)
```

```python
###columns that contain the word uninsured
df.columns[df.columns.str.contains('uninsured')]
```

## EDA 1
#### Scatter plot of uninsured against population

```python
sns.regplot(data=df[['population','uninsured']],x='population',y='uninsured', fit_reg=False, color="teal")
```

# Observation
#### 1. Uninsured rate is generally less than 35% (0.35)
#### 2. Smaller populations(<5 million) have the highest uninsurance

# EDA 2
### Features linked to uninsurance (by correlation). Top 10.

```python
###getting dataframe of top 10 factors.
factor_df=df.drop(columns=['name','uninsured','uninsured_adults','uninsured_children'])
df_corr=factor_df.corrwith(other=df.uninsured).rename('corr_values').to_frame()
top_10=df_corr[df_corr.corr_values>0].sort_values(by='corr_values', ascending=False).head(10)
top_10
```

# Observation
#### Factors most correlated to uninsured can be grouped into:
#### Poverty
#### Teen births 
#### Hispanics 
#### Motor Vehicle related incidences 
#### Health related issues

```python
top_5_df = df[['poverty','teen_births','children_eligible_for_free_or_reduced_price_lunch',
               'off_road_motor_vehicle_crash_related_er_visits','%_hispanic','uninsured']]
```

```python
top_5_df.plot.scatter(x="uninsured", y="%_hispanic",color='teal');
```

```python
#as the number of teen births increases, the number of uninsured increases
top_5_df.plot.scatter(x="uninsured", y="teen_births",color='teal');
```

```python
top_5_df.plot.scatter(x="uninsured", y="poverty",color='teal');
```

```python
sns.pairplot(top_5_df, kind='scatter',)
```

# EDA 3
#### Countys with the highest uninsurance rate. Top 10.

```python
top_10_uninsured_countys=df[['name','uninsured']].groupby('name').mean().sort_values(by='uninsured', ascending=False).head(10)
top_10_uninsured_countys
```

```python
#top 10 counties with the heighest uninsured rate
top_10_uninsured_countys.plot.bar(stacked=False, color='teal');
```
